### Principal Component Analysis (PCP) Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Loadd dataset
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

```python
## apllying scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
df
```

```python
scaled_data = scaler.transform(df)
```

```python
pd.DataFrame(scaled_data)
```

```python
## Applying PCA 
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
data_pca = pca.fit_transform(scaled_data)
print(sum(pca.explained_variance_ratio_))
```

```python
data_pca
```

```python
pca.explained_variance_
# 13.30 → PC1 mein kitna "spread" / information hai
# 5.70 → PC2 mein kitna "spread" / information hai
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(data_pca[:,0], data_pca[:,1], c=cancer_dataset['target'], cmap='plasma')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA - Breast Cancer Dataset")
plt.colorbar(scatter, label='Target (0=Malignant, 1=Benign)')
plt.show()
```
