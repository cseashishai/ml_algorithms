## Logicstic Regression with Roc Curv and Roc Auc Score

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
X,y= make_classification(n_samples=1000,n_classes=2,random_state=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
## Creat a dummy model with default 0 as output
dummy_model_prob=[0 for _ in range(len(y_test ))]
```

```python
dummy_model_prob
```

```python
### Lets creat basic logicstic modle
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
## Predection based on the predection
model_prob = model.predict_proba(X_test)
print(model_prob.shape)
model_prob = model_prob[:, 1]## lets foucuse on the positive outcomes
```

```python
## lets calculate the score
dummy_model_auc=roc_auc_score(y_test,dummy_model_prob)
model_auc=roc_auc_score(y_test,model_prob)
print(dummy_model_auc)
print(model_auc)
```

```python
## Calculate Roc Curve
dummy_fpr ,dummy_tpr,_ = roc_curve(y_test,dummy_model_prob)
model_fpr,model_tpr,thresholds =roc_curve(y_test,model_prob)
```

```python
thresholds
```

```python
from matplotlib import pyplot
```

```python
model_fpr
```

```python
# plot the curve for the model 
pyplot.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
pyplot.plot(model_fpr,model_tpr,linestyle='dotted',label='Logicstic')

# axis label
pyplot.xlabel('False positive rate')
pyplot.ylabel('True Positive rate')
pyplot.legend()
pyplot.show()
```

```python
# plot the curve for the model 
import numpy as np
fig =pyplot.figure(figsize=(20,50))
pyplot.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
pyplot.plot(model_fpr, model_tpr, marker='.', label='Logistic')
ax =fig.add_subplot(111)
for xyz in zip(model_fpr,model_tpr,thresholds):
    ax.annotate('%s'% np.round(xyz[2],2), xy=(xyz[0],xyz[1]))

# axis label
pyplot.xlabel('False positive rate')
pyplot.ylabel('True Positive rate')
pyplot.legend()
pyplot.show()
```
