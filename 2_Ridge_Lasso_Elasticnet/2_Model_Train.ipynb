```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_cleaned_dataet.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['month','year','day'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Feature Encoding 
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## Dependent and independed features
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X.head()
```

```python
y
```

```python
# Train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Featuer Selection based of correletion
X_train.corr()
```

```python
## Chek for Multi colleniraty
plt.Figure(figsize=(30,10))
corr = X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
X_train.shape,X_test.shape
```

```python
def correletion(dataset,threshold):
    col_corr=set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr            
```

```python
## Threshold -Domain Expertise
corr_feature = correletion(X_train,0.85)
```

```python
# Drop the Features when correletion is more than 0.85
X_train.drop(corr_feature,axis=1,inplace=True)
X_test.drop(corr_feature,axis=1,inplace=True)
X_train.shape,X_test.shape
```

```python
## Feture Scaling or standerization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
## Box Plot to understand effect of standard scaler
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_Train before scaling')
plt.subplot(1,2,2)
sns.boxenplot(data=X_train_scaled)
plt.title('X_test_scaled After Scaled')
```

### Linner Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
linreg = LinearRegression()
```

```python
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mae= mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean absoulute error ",mae)
print("R2 Score",score)
plt.scatter(y_test,y_pred)
```
