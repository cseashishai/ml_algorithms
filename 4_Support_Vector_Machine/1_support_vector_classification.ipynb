### Support Vector Classifire Implementation

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
 ## Lets Creat Synthetic data points
from sklearn.datasets import make_classification
# from sklearn.decomposition import PCA
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,
                        n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
# sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
svc=SVC(kernel='linear') # when we use for data are simple strate line
```

```python
svc.fit(X_train,y_train)        
```

```python
# Predected
y_pred =svc.predict(X_test)
```

```python
svc.coef_
```

```python
svc.intercept_
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf =SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
y_pred1=rbf.predict(X_test)
```

```python
y_pred1
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynimial=SVC(kernel='poly')
polynimial.fit(X_train,y_train)

## Predection
y_pred2=polynimial.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)

## Predection
y_pred3=sigmoid.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid.intercept_
```

## Hyperparametere Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

## Define parameter 
pram_grid ={
    'C':[0.1,1,10,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}
```

```python
grid=GridSearchCV(SVC(),param_grid=pram_grid,cv=5,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
 
```
